```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
from scipy import stats
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib notebook
year_avg = df.mean(axis = 1)
year_std = df.std(axis = 1)
yerr = year_std/np.sqrt(df.shape[1])* stats.t.ppf(.95, df.shape[1]-1)
data = range(df.shape[0])
plt.figure()
colors = ["red", "tan", "lightblue", "blue"]
bars = plt.bar(data, year_avg, yerr = yerr, alpha = .6, color = colors)
thresh = 42000
plt.axhline(y = thresh, color = "grey", alpha = .5)
plt.xticks(range(df.shape[0]), ["1992","1993","1994","1995"], alpha = 1)
cmap = mpl.cm.RdBu
norm = mpl.colors.Normalize(vmin = 0, vmax = 1)
plt.colorbar(mpl.cm.ScalarMappable(norm=norm, cmap=cmap), orientation = "horizontal")
```
